# Step 5:Complete the fundamental data quality and consistency checks 

```python
## check for mixed types
for col in df_cus.columns.tolist():
  weird = (df_cus[[col]].applymap(type) != df_cus[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cus[weird]) > 0:
    print (col)
```

#### user_id is the column common in both dataframes and its datatype is same. So I merging both dataframes with inner join and user_id as key column

```python
# merging both dataframes
df_final_merge = df_prepared.merge(df_cus, on = 'user_id', indicator = True)
```

```python
# checking the status of merge
df_final_merge['_merge'].value_counts(dropna=False)
```

#### All entries have value both in merge indicator. So its a full match.

```python
#Dropping the '_merge' column 
df_final_merge = df_final_merge.drop(columns = '_merge')
```
