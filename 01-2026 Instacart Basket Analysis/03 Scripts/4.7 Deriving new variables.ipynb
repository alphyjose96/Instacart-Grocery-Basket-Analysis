```python
# Defining function to categorize products based on price range
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Applying the function
df['price_range'] = df.apply(price_label, axis=1)
```

# Step 2: Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
#function to create a column for busiest days
result_1 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_1.append("Busiest days")
  elif value == 4 or value == 3:
    result_1.append("Slowest days")
  else:
    result_1.append("Regularly busy")
```

```python
##Creating a new column and seting the result of the function to that column
df_ords_prods_merge['busiest_days'] = result_1
```

# Step 4: When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# checking the frequency of the column
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

```python
# creating  a new function to split the hours in to three labels, whicha are
# Most orders = 10, 11, 14, 15, 13, 12, 16, and 9
# Fewest orders =  23, 6, 0, 1, 5, 2, 4, and 3
# Average orders = All the remaining hours
result_2 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]: 
   result_2.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_2.append("Fewest orders")
  else: result_2.append("Average orders")
```

```python
# Creating new column busiest_period_of_day and assigining the result to it
df_ords_prods_merge['busiest_period_of_day'] = result_2
```

```python
# checking the frequency of the column
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna=False)
```
