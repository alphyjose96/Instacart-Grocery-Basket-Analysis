# Step 1: Merge your prepared Instacart orders data with the new orders_products_prior dataframe

```python
# merging both dataframes with outer join
df_merged_test = df_ords.merge(df_ords_prior, on = 'order_id',how = 'outer' ,indicator = True)
```
