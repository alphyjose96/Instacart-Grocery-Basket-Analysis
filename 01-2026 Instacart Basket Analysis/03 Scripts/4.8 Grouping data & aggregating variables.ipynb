# Step 4: Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
#using the transform() function to create a new column containing the maximum frequency of the “order_number”
df_orders_products_merged_derived['max_order'] = df_orders_products_merged_derived.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_orders_products_merged_derived.head(15)
```

```python
# creating new column that assigns a loyalty flag to each customer according to the criteria (via the loc() function)
```

```python
df_orders_products_merged_derived.loc[df_orders_products_merged_derived['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_orders_products_merged_derived.loc[(df_orders_products_merged_derived['max_order'] <= 40) & (df_orders_products_merged_derived['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_orders_products_merged_derived.loc[df_orders_products_merged_derived['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_orders_products_merged_derived[['user_id', 'loyalty_flag', 'order_number','max_order']].head(60)
```

```python
# printing frequency of loyalty flags
df_orders_products_merged_derived['loyalty_flag'].value_counts(dropna=False)
```

# Step 5: The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers

```python
# the min, max and mean of product prices of  each loyalty category
df_orders_products_merged_derived.groupby(['loyalty_flag']).agg({'prices':['mean','min','max']})
```

#### There isnt much significant difference between the spending habits of three types of customers.

# Step 6: Create a spending flag for each user based on the average price across all their orders

```python
#using the transform() function to create a new column containing the average spend made by a user
df_orders_products_merged_derived['average_spend'] = df_orders_products_merged_derived.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# # creating new column that assigns a flag to each customer according to the average (via the loc() function)
```

```python
df_orders_products_merged_derived.loc[df_orders_products_merged_derived['average_spend'] < 10, 'spend_flag'] = 'Low spender'
```

```python
df_orders_products_merged_derived.loc[df_orders_products_merged_derived['average_spend'] >= 10, 'spend_flag'] = 'High spender'
```

```python
# printing frequency of spend flags
df_orders_products_merged_derived['spend_flag'].value_counts(dropna=False)
```

#### More number of users purchase products  less than $10 

# Step 7: Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. 

```python
#using the transform() function to create a new column containing the median of 'days_since_prior_order_column'
df_orders_products_merged_derived['order_frequency'] = df_orders_products_merged_derived.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#creating new column that assigns a flag to each customer according to the order frequency (via the loc() function)
```

```python
df_orders_products_merged_derived.loc[df_orders_products_merged_derived['order_frequency'] > 10 & (df_orders_products_merged_derived['order_frequency'] <=20) , 'frequency_flag'] = 'Regular customer'
```

```python
df_orders_products_merged_derived.loc[df_orders_products_merged_derived['order_frequency'] <= 10 , 'frequency_flag'] = 'Frequent customer'
```

```python
df_orders_products_merged_derived.loc[df_orders_products_merged_derived['order_frequency'] > 20 , 'frequency_flag'] = 'Non-frequent customer'
```

```python
# printing frequency of spend flags
df_orders_products_merged_derived['frequency_flag'].value_counts(dropna=False)
```

#### Instacart has higher number of frequent customers. And there are 5 first time customers wich is indicated as NaN
