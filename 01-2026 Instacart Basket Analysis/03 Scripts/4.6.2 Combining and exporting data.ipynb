# Step 5:Determine a suitable way to combine the orders_products_combined dataframe with your products data set.

```python
#import the product dataframe from prepared data file.
df_prds= pd.read_csv(os.path.join(path, '02 Data','Prepared Data', 'products_checked.csv'),index_col=0)
```

```python
# Test merge without overwriting
pd.merge(df_ords_prds_comb,df_prds, on = 'product_id',indicator = True)
```

```python
# dropping _merge column from df_ords_prds_comb dataframe because merging will create a new indicator column with same name and its is showing and error
df_ords_prds_comb = df_ords_prds_comb.drop(['_merge'], axis=1)
```

```python
df_ords_prds_comb.head()
```

```python
# Test merge without overwriting
pd.merge(df_ords_prds_comb,df_prds, on = 'product_id',indicator = True)
```

```python
# merging both dataframes to create df_ords_prods_merge. I am not explicitly providing the 'how' argument because the 
# default join is inner join.
df_ords_prods_merge= df_ords_prds_comb.merge(df_prds, on = 'product_id', indicator = True)
```

#### The order of DataFrames doesnt makes a difference in the resulting dataframe when using the merge function in pandas. 
#### The core data (which rows are kept) should be consistent regardless of order.
#### The difference is only in terms of the column order in the output DataFrame and the handling of overlapping columns not used as keys. 
#### I am gvining df_ords_prds_comb dataframe first , so that the resulting dataframe will have order details first
#### And there are no columns with same name other than the key in both dataframes.
